#### 过拟合与模型选择

这里我们使用 $ y = sin2\pi x$ 加上一个正态分布的噪声干扰，用多项式去拟合

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.optimize import  leastsq
```

```python
# 目标函数
def target_func(x):
    return np.sin(2 * np.pi * x)

# 多项式 p: 多项式参数， x输入值
def guantic_func(p, x):
    f = np.poly1d(p)
    return f(x)

# 残差 预测值和真实值之间的差异
def res_func(p, x, y):
    ret = guantic_func(p, x) - y 
    return ret
```

```python
x = np.linspace(0, 1, 10) # 1-10之间均匀选择10个点
x_points = np.linspace(0, 1, 1000) # 用于画曲线
# 加上正态分布噪声的输出值
real_y = target_func(x)
y = [np.random.normal(0, 0.1) + y1 for y1 in real_y]
```

```python
# 拟合 M 为多项式的次数
def fitting(M = 0):
    p_init = np.random.rand(M+1) #随机初始化多项式的参数
    # 最小二乘法
    p_lsq = leastsq(res_func, p_init, args=(x, y))
    
    # 可视化
    plt.plot(x_points, target_func(x_points), label="Real")
    plt.plot(x_points, guantic_func(p_lsq[0], x_points), label="Fitted")
    plt.scatter(x, y, c='r', label="Noise")
    
    plt.legend()

    
fitting(M=0)
```

```python
fitting(M=1)
```

```python
fitting(M=3)
```

```python
fitting(M=4)
```

```python
fitting(M=9)
```
